```python
from __future__ import print_function

# Import conventions we'll be using here. See Part 1
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Plot some datasets
* Plot them individualy
* Plot in one graph
  * Scale is different then we use log to lower the huge difference.
  

```python
datafile = open('DATAgg_d4.txt','r')
data =  datafile.readlines() 
datafile.close()
a = []
b = []
for row in data:
    this_data = row.split()
    a.append(float(this_data[0]))
    b.append(float(this_data[1]))
a = np.array(a)
b = np.array(b)
fig,ax = plt.subplots()
ax.plot(a[:100],b[:100]*10**3,label='.') # turn pb into fb
ax.set( xlabel='m', ylabel='fb')
plt.show()
```

```python
datafile = open('DATAgg_d4.txt','r')
data =  datafile.readlines()
datafile.close()

s = []
cs = []
for row in data:
    this_data = row.split()
    s.append(float(this_data[0]))
    cs.append(float(this_data[1]))


S = np.array(s)
CS = np.array(cs)

plt.plot(S,CS*10**3,'.g')
plt.xlabel("$M_D (MeV)$ ")
plt.ylabel("$\sigma$ (fb)")
plt.title('Cross section with respect to S with  $\delta = 4 $')
plt.show()          
```

```python
datafile = open('DATAqq_d4.txt','r')
data =  datafile.readlines()
datafile.close()

s = []
cs = []
for row in data:
    this_data = row.split()
    s.append(float(this_data[0]))
    cs.append(float(this_data[1]))


S = np.array(s)
CS = np.array(cs)

plt.plot(S,CS*10**3,'.g')
plt.xlabel("$M_D (MeV)$ ")
plt.ylabel("$\sigma$ (fb)")
plt.title('Cross section  with respect to $M_D$ with  $\delta = 4 $')
plt.show()          
```
